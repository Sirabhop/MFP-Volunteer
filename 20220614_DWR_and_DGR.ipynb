#**Step 2: Look in to the ministry of natural resources and environment**

**Filter only Ministry of NE and E**

```python
df_nr = df[df['MINISTRY'] == "กระทรวงทรัพยากรธรรมชาติและสิ่งแวดล้อม"]
```

**Group budgetary_unit according to organization chart**

```python
public_org_name = ['สำนักงานปลัดกระทรวงทรัพยากรธรรมชาติและสิ่งแวดล้อม',
                   'สำนักงานนโยบายและแผนทรัพยากรธรรมชาติและสิ่งแวดล้อม',
                   'กรมควบคุมมลพิษ',
                   'กรมส่งเสริมคุณภาพสิ่งแวดล้อม',
                   'กรมทรัพยากรน้ำ',
                   'กรมทรัพยากรน้ำบาดาล',
                   'กรมอุทยานแห่งชาติ สัตว์ป่า และพันธุ์พืช',
                   'กรมป่าไม้',
                   'กรมทรัพยากรทางทะเลและชายฝั่ง',
                   'กรมทรัพยากรธรณี']
gov_owned_org_name = ['องค์การอุตสาหกรรมป่าไม้',
                      'องค์การสวนสัตว์แห่งประเทศไทย',
                      'องค์การสวนพฤกษศาสตร์']
private_org_name = ['องค์การบริหารจัดการก๊าซเรือนกระจก (องค์การมหาชน)',
                    'สำนักงานพัฒนาเศรษฐกิจจากฐานชีวภาพ (องค์การมหาชน)']
```

```python
#df_nr.insert(4, 'ORG_NAME', 'Null')
df_nr.loc[df_nr['BUDGETARY_UNIT'].isin(public_org_name), 'ORG_NAME'] = 'ส่วนราชการ'
df_nr.loc[df_nr['BUDGETARY_UNIT'].isin(gov_owned_org_name), 'ORG_NAME'] = 'รัฐวิสาหกิจ'
df_nr.loc[df_nr['BUDGETARY_UNIT'].isin(private_org_name), 'ORG_NAME'] = 'องค์การมหาชน'
```

```python
df_nr_water = df_nr[(df_nr['BUDGETARY_UNIT'] == 'กรมทรัพยากรน้ำ') | (df_nr['BUDGETARY_UNIT'] == 'กรมทรัพยากรน้ำบาดาล')]
```

```python
df_nr_water.info()
```

**Text extraction**

```python
#create new columns
df_nr_water.insert(17, 'TYPE', np.NaN)
df_nr_water.insert(18, 'PROVINCE', np.NaN)
df_nr_water.insert(19, 'MAIN_TYPE', np.NaN)
```

```python
#Combinded those budget category
df_nr_water.loc[df_nr_water['TYPE'].isna(), 'TYPE'] = df_nr_water.loc[df_nr_water['TYPE'].isna(), 'CATEGORY_LV4']
df_nr_water.loc[df_nr_water['TYPE'].isna(), 'TYPE'] = df_nr_water.loc[df_nr_water['TYPE'].isna(), 'CATEGORY_LV3']
df_nr_water.loc[df_nr_water['TYPE'].isna(), 'TYPE'] = df_nr_water.loc[df_nr_water['TYPE'].isna(), 'CATEGORY_LV2']
df_nr_water.loc[df_nr_water['TYPE'].isna(), 'TYPE'] = df_nr_water.loc[df_nr_water['TYPE'].isna(), 'CATEGORY_LV1']
```

```python
#Import province
province = pd.read_excel('25640531_sc002_.xlsx', usecols = ['ProvinceNameThai', 'RegionDGR', 'Region_WaterResourceOffice', 'Region_RDPB', 'RiverBasin'])
province.rename(columns = {'ProvinceNameThai': 'PROVINCE'}, inplace = True)
```

*Remark: province info is provided by DGA ([link](https://data.go.th/dataset/proviceandregionthailand))*

```python
#Type 1: full province
province_type1_list = province.PROVINCE.to_list()
#Type 2: in case the item_description doesn't contain จังหวัด
province_type2_list = []
for i in range(len(province_type1_list)):
  province_type2_list.append(province_type1_list[i].replace('จังหวัด', ''))
```

```python
for i in range(len(province_type2_list)):
  df_nr_water.loc[df_nr_water.ITEM_DESCRIPTION.str.contains(province_type2_list[i]), 'PROVINCE'] = province_type1_list[i]
```

```python
df_nr_water = pd.merge(df_nr_water, province, on = 'PROVINCE', how = 'left')
```

```python
#group higher-level category and saved to MAIN_TYPE
MAIN_TYPE = []
for row in df_nr_water['TYPE']:
  if row in ['ค่าปรับปรุงแหล่งน้ำ', 'ค่าปรับปรุงสิ่งก่อสร้างอื่น ๆ']:
    MAIN_TYPE.append('ค่าปรับปรุงแหล่งน้ำและสิ่งก่อสร้าง')
  elif row in ['ค่าก่อสร้างแหล่งน้ำ', 'ค่าสำรวจออกแบบ', 'ค่าควบคุมงาน', 'ค่าที่ดิน/สิ่งก่อสร้าง']:
    MAIN_TYPE.append('ค่าก่อสร้างแหล่งน้ำ')
  elif row in ['งบบุคลากร', 'เงินเดือนและค่าจ้างประจำ']:
    MAIN_TYPE.append('งบบุคลากร')
  elif row in ['ค่าครุภัณฑ์', 'ครุภัณฑ์การเกษตร', 'ครุภัณฑ์นวัตกรรมไทย - ด้านการเกษตร', 'ค่าตอบแทนใช้สอยและวัสดุ', 'ครุภัณฑ์วิทยาศาสตร์', 'ครุภัณฑ์ยานพาหนะและขนส่ง', 'ครุภัณฑ์คอมพิวเตอร์', 'ครุภัณฑ์สำรวจ']:
    MAIN_TYPE.append('ค่าครุภัณฑ์')
  else:
    MAIN_TYPE.append('งบรายจ่ายอื่น')
df_nr_water['MAIN_TYPE'] = MAIN_TYPE
```

```python
#Look up for the given budgetary unit
for i in range(len(province.Region_WaterResourceOffice)):
  df_nr_water.loc[df_nr_water.ITEM_DESCRIPTION.str.contains(province.Region_WaterResourceOffice[i]), 'RECEIVED_WRO'] = province.Region_WaterResourceOffice[i]
for i in range(len(province.RegionDGR)):
  df_nr_water.loc[df_nr_water.ITEM_DESCRIPTION.str.contains(province.RegionDGR[i]), 'RECEIVED_DGR'] = province.RegionDGR[i]
```
